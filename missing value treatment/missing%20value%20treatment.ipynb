# filling the missing values-IMPUTATION

```python
#to fill the missing value in the table eith the zero
df.fillna(0)
```

```python
#to fill the missing value in a particulare column
df["Age"].fillna(0)
```

```python
#to fill the missing value with the mean age
df["Age"].fillna(df["Age"].mean())
```
